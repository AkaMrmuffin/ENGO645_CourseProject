```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
```

```python
mypath1 = r'D:\AltFemp\raw_data\minning_data'
files = [f for f in listdir(mypath1) if isfile(join(mypath1, f))]
```

```python
i = 0 
for f in files:
    name = f[0:-3]
    df = pd.read_csv(r'D:\AltFemp\raw_data\minning_data\%s'%f,sep=',')
    dim = df.shape
    r = df.ch4_recordtime
    range = pd.date_range(r[0], r[dim[0]-1], periods=dim[0])
    df.index =range
    second_sum = pd.DataFrame()
    second_sum['estimated_heading'] = df.estimated_heading.resample('S').mean()
    second_sum['true_wspd'] = df.true_wspd.resample('S').mean()
    second_sum['ch4_diag'] = df.ch4_diag.resample('S').mean()
    second_sum['ch4_ch4d'] = df.ch4_ch4d.resample('S').mean()
    second_sum['ch4_rssi'] = df.ch4_rssi.resample('S').mean()
    second_sum['altitude'] = df.altitude.resample('S').mean()
    second_sum['ch4_sonic_temp'] = df.ch4_sonic_temp.resample('S').mean()
    second_sum['ch4_wdir_cf'] = df.ch4_wdir_cf.resample('S').mean()
    second_sum['ch4_moleratio'] = df.ch4_moleratio.resample('S').mean()
    second_sum['ch4_temp'] = df.ch4_temp.resample('S').mean()
    second_sum['ch4_pressure'] = df.ch4_pressure.resample('S').mean()
    second_sum['spd_over_grnd'] = df.spd_over_grnd.resample('S').mean()
    second_sum['ch4_mr_anomaly'] = df.ch4_mr_anomaly.resample('S').mean()
    second_sum['ch4_mr_mean'] = df.ch4_mr_mean.resample('S').mean()
    second_sum['ch4_mr_median'] = df.ch4_mr_median.resample('S').mean()
    second_sum['ch4_mr_sd'] = df.ch4_mr_sd.resample('S').mean()
    second_sum['true_track'] = df.true_track.resample('S').mean()
    second_sum['x_utm'] = df.x_utm.resample('S').mean()
    second_sum['y_utm'] = df.y_utm.resample('S').mean()
    second_sum['wdir_cf_corrected'] = df.wdir_cf_corrected.resample('S').mean()
    second_sum['wspd_cf_corrected'] = df.wspd_cf_corrected.resample('S').mean()
    second_sum['true_wdir'] = df.true_wdir.resample('S').mean()
    second_sum['pitch'] = df.pitch.resample('S').mean()
    second_sum['roll'] = df.roll.resample('S').mean()

    
    
    second_sum.to_excel(r'D:\AltFemp\new_downsample_Data\%s'%name+"xlsx")
    print (i)
    i = i + 1 
```
